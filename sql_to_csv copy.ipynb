```python
from sqlalchemy import create_engine
from sqlalchemy.sql import text
import MySQLdb
import pandas as pd
import numpy as np
import pandas.io.sql as psql
import json
```

```python
global teamnamedict
# C:\Users\mg\JupyterLabDir\Rest\Pr Winning\teamnamedict_streamlit.jso
with open('teamnamedict_streamlit.json') as f:
    teamnamedict = json.load(f)
```

```python
def process_team_names_of_df(x_df):
    x_df = x_df.replace(teamnamedict)
    return x_df
```

```python
outputname = "serie_a2021__14022022.csv"
```

```python
# setup the database connection.  There's no need to setup cursors with pandas psql.
db = MySQLdb.connect(host='127.0.0.1',
                        database='xg',
                        user='root',
                        password='root')

xgtablename = "serie_a2021__14022022"

# create query for all users
queryxg = "select * from {0} ".format(
    xgtablename)

print("queryxg", queryxg)

# execute the query and assign it to a pandas dataframe
dfxg = psql.read_sql(queryxg, con=db)
dfxg = process_team_names_of_df(dfxg)

# close the database connection
db.close()
```
